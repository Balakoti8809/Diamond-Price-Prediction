```python
# sector analysis
import re
# Group by 'sector' and calculate the average price
avg_price_per_Clarity = df.groupby('Clarity')['Price(in US dollars)'].mean().reset_index()

# Function to extract sector numbers
def extract_Clarity_value(Clarity_value):
    match = re.search(r'\d+', Clarity_value)
    if match:
        return int(match.group())
    else:
        return float('inf')  # Return a large number for non-numbered sectors

avg_price_per_Clarity['Clarity_value'] = avg_price_per_Clarity['Clarity'].apply(extract_Clarity_value)

# Sort by sector number
avg_price_per_Clarity_sorted_by_Clarity = avg_price_per_Clarity.sort_values(by='Clarity_value')

# Plot the heatmap
plt.figure(figsize=(5,8))
sns.heatmap(avg_price_per_Clarity_sorted_by_Clarity.set_index('Clarity')[['Price(in US dollars)']], annot=True, fmt=".2f", linewidths=.5)
plt.title('Average Price per Clarity (Sorted by Clarity Value)')
plt.xlabel('Average Price')
plt.ylabel('Clarity')
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df_num.corr()

# Plot the heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,        
    fmt=".2f",          
    cmap="coolwarm",    
    vmin=-1, vmax=1,    
    square=True,        
    cbar_kws={'shrink': 0.8}  
)

# Add title and axis labels
plt.title("Heatmap of Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha="right")  
plt.yticks(rotation=0)
plt.tight_layout()  
plt.show()
```
